## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```
# (A) = valor_total_de_vendas
qtd_total_vendas = 3
ticket_medio = 320.52

(A) = qtd_total_vendas*ticket_medio 
print (A)
```

```
# (B) = qtd_total_vendas
valor_total_de_vendas = 834.47
ticket_medio = 119.21

(B) = valor_total_de_vendas // ticket_medio

valor_total_de_vendas = (B) * ticket_medio
print (B)
```

```
# (C) = ticket_medio
valor_total_de_vendas = 15378.12
qtd_total_vendas = 5

(C) = valor_total_de_vendas/qtd_total_vendas
print (C)
```

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'
x = cancao.count("roda")
print (x)

y = cancao.split(',')
print (y)

z = cancao.lower()
print (z)
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'
posicao_div = noticia.find('2')
print (posicao_div)
```

```
selic = noticia[posicao_div+0:17]
print (selic)
```

```
posicao_ano = noticia.find('6')
print (posicao_ano)
```

```
ano = noticia[posicao_ano+0:63]
print (ano)
```
